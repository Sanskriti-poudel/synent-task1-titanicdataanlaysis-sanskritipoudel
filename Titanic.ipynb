### Handle missing values
### Remove duplicates if present


```python
print("\n Missing Values")
df.isnull().sum()
```

```python
print("\n Duplicate Values")
df.duplicated().sum()
```

```python
#Filling Null Values for Embarked Column
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
#Handling Missing Values for Cabin Column
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
#Filling Null Values for Age
#Getting the title form name column
df["Name"]
```

```python
s='Braund, Mr. Owen Harris'
a=s.split(',')
a
```

```python
b=s.split(',')[1]
b
```

```python
c=b.strip()
c
```

```python
d=c.split('.')[0]
d
```

```python
df['Title']=df['Name'].str.split(',').str[1].str.split('.').str[0].str.strip()
```

```python
df.head()
```

```python
group_using_title=df.groupby('Title')['Age'].mean().round(2)
group_using_title
```

```python
df["Age"]=df["Age"].fillna(df["Title"].map(group_using_title))
```

```python
df.isnull().sum()
```

### Convert Data Types

```python
# Check current data types
print(df.dtypes)
```

```python
df['Survived'] = df['Survived'].astype('category')
df['Age'] = df['Age'].astype(int)
df['Fare'] = df['Fare'].astype(int)
```

```python
df.head(3)
```

### Rename columns

```python
df.rename(columns={
    'PassengerId': 'ID',
    'Survived': 'Survival_Status',
    'Pclass': 'Ticket_Class',
    'Name': 'Passenger_Name',
    'Sex' : 'Gender'
}, inplace=True)
```
